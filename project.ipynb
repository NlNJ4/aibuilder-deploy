```python
from sentence_transformers import SentenceTransformer, InputExample, losses, evaluation, util , SentencesDataset
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator
from torch.utils.data import DataLoader
import torch
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
df = pd.read_csv("/kaggle/input/deploy/data.csv")
```

```python
q = df['query']
desc = df['Description']
menu = df['menu']
```

```python
for i,val in enumerate(q):
    if "." in val:
        q[i] = val.split(".")[1].strip()
    elif "*" in val:
        q[i] = val.split("*")[1].strip()
    elif "-" in val:
        q[i] = val.split("-")[1].strip()
```

```python
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
embeddings = model.encode(desc, convert_to_tensor=True)
```

```python
score = 0
for i in range(len(q)):
    clear_output(wait=True)
    query_en = model.encode(q[i],convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_en, embeddings)
    all_idx = torch.topk(cosine_scores.flatten(), 5).indices
    if i in all_idx:
        score = score + 1
```

```python
score/len(q)
```

```python
clear_output(wait=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(q, desc, test_size=0.25, random_state=42)
```

```python
from sentence_transformers import InputExample

# Assuming x_train and y_train are pandas Series or lists
if len(x_train) != len(y_train):
    raise ValueError("x_train and y_train must be of the same length")

traindata = []

for i in range(len(x_train)):
    try:
        x = str(x_train[i])
        y = str(y_train[i])
        traindata.append(InputExample(texts=[x, y], label=1.0))
    except KeyError as e:
        print(f"KeyError at index {i}: {e}")
        continue
    except Exception as e:
        print(f"Unexpected error at index {i}: {e}")
        continue

# Verify the resulting list
print(traindata)
```

```python
x_train = x_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
```

```python
len(x_train)
```

```python
traindata = []
for i in range(len(x_train)):
  traindata.append(InputExample(texts=[str(x_train[i]), str(y_train[i])], label = 1.0))
```

```python
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
```

```python
train_dataloader = DataLoader(traindata, shuffle=True, batch_size=16)
```

```python
train_loss = losses.MultipleNegativesRankingLoss(model=model)
```

```python
model.fit(train_objectives=[(train_dataloader, train_loss)],epochs=20, warmup_steps=100, show_progress_bar=True)
```

```python
embeddings = model.encode(desc, convert_to_tensor=True)
```

```python
x_test = x_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
score = 0
for i in range(len(x_test)):
    ans = []
    query_en = model.encode(x_test.iloc[i], convert_to_tensor=True)  # ใช้ iloc เพื่อเข้าถึงแถว
    cosine_scores = util.cos_sim(query_en, embeddings)
    all_idx = torch.topk(cosine_scores.flatten(), 5).indices
    for j in all_idx:
        ans.append(desc[int(j)])  # แปลง tensor เป็น int
    if y_test.iloc[i] in ans:  # ตรวจสอบ y_test ของรายการที่เกี่ยวข้อง
        score += 1

print(f'Score: {score/len(x_test)}')
```
